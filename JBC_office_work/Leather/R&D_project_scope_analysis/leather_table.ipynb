# chats table

```python
import pandas as pd
import pymysql
from googletrans import Translator

# ğŸ“Œ User-defined: columns to translate
columns_to_translate = {
    "message": "message_jap"

}

# ğŸ“Œ Optional: set this to None to auto-use all rows
num_rows_to_translate = 50  

# ğŸŒ Translator init
translator = Translator()

# ğŸ” DB config
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': '',
    'database': 'Leather_repair_db',
    'port': 3306
}

# ğŸ§¾ SQL query
query = "SELECT * FROM chats"

# ğŸ“¥ Fetch data from DB
try:
    connection = pymysql.connect(**db_config)
    df = pd.read_sql(query, connection)
    connection.close()
    print("âœ… Data fetched successfully!")
except Exception as e:
    print("âŒ Error while connecting or fetching:", e)

# ğŸ” Set number of rows to translate
if num_rows_to_translate is None:
    num_rows_to_translate = len(df)
    print(f"ğŸ“Š No row limit set. Translating all {num_rows_to_translate} rows.")

# ğŸŒ Translate selected columns
for original_col, new_col in columns_to_translate.items():
    if original_col in df.columns:
        try:
            print(f"ğŸ” Translating first {num_rows_to_translate} rows of '{original_col}' â¡ï¸ '{new_col}' ...")
            translated_col = []

            for i, text in enumerate(df[original_col].astype(str)):
                if i < num_rows_to_translate:
                    try:
                        translated_text = translator.translate(text, src='ja', dest='en').text
                    except Exception as trans_err:
                        translated_text = "Translation Failed"
                else:
                    translated_text = ""
                translated_col.append(translated_text)

            df[new_col] = translated_col

        except Exception as e:
            print(f"âŒ Error translating '{original_col}': {e}")
            df[new_col] = "Translation Failed"
    else:
        print(f"âš ï¸ Column '{original_col}' not found in DataFrame.")

# ğŸ’¾ Save updated DataFrame
output_file = "chats_translated.xlsx"
try:
    df.to_excel(output_file, index=False)
    print(f"ğŸ“ Excel file saved successfully as '{output_file}'")
except Exception as e:
    print("âŒ Error while saving to Excel:", e)
```

```python
import pandas as pd
import pymysql
from googletrans import Translator

# ğŸ“Œ User-defined: columns to translate
columns_to_translate = {
    "internal_memo": "internal_memo_translated",
    "customer_memo": "customer_memo_translated"
}

# ğŸ“Œ Optional: set this to None to auto-use all rows
num_rows_to_translate = 20

# ğŸŒ Translator init
translator = Translator()

# ğŸ” DB config
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': '',
    'database': 'Leather_repair_db',
    'port': 3306
}

# ğŸ§¾ SQL query
query = "SELECT * FROM orders"

# ğŸ“¥ Fetch data from DB
try:
    connection = pymysql.connect(**db_config)
    df = pd.read_sql(query, connection)
    connection.close()
    print("âœ… Data fetched successfully!")
except Exception as e:
    print("âŒ Error while connecting or fetching:", e)

# ğŸ” Set number of rows to translate
if num_rows_to_translate is None:
    num_rows_to_translate = len(df)
    print(f"ğŸ“Š No row limit set. Translating all {num_rows_to_translate} rows.")

# ğŸŒ Translate selected columns
for original_col, new_col in columns_to_translate.items():
    if original_col in df.columns:
        try:
            print(f"ğŸ” Translating first {num_rows_to_translate} rows of '{original_col}' â¡ï¸ '{new_col}' ...")
            translated_col = []

            for i, text in enumerate(df[original_col].astype(str)):
                if i < num_rows_to_translate:
                    try:
                        translated_text = translator.translate(text, src='ja', dest='en').text
                    except Exception as trans_err:
                        translated_text = "Translation Failed"
                else:
                    translated_text = ""
                translated_col.append(translated_text)

            df[new_col] = translated_col

        except Exception as e:
            print(f"âŒ Error translating '{original_col}': {e}")
            df[new_col] = "Translation Failed"
    else:
        print(f"âš ï¸ Column '{original_col}' not found in DataFrame.")

# ğŸ’¾ Save updated DataFrame
output_file = "orders.xlsx"
try:
    df.to_excel(output_file, index=False)
    print(f"ğŸ“ Excel file saved successfully as '{output_file}'")
except Exception as e: 
    print("âŒ Error while saving to Excel:", e)
```
